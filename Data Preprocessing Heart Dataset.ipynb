```python
## Categorical Col
categorical_columns=df.select_dtypes(include='object').columns
print(categorical_columns)
```

```python
 l_e=LabelEncoder()
```

```python
## Label Encoding
for col in categorical_columns:
    df[col]=l_e.fit_transform(df[col])
```

```python
df.tail(50)
```

```python
## Replace 0 cholesterol with mean
df['Cholesterol'].replace(0,df['Cholesterol'].mean(),inplace=True)
```

```python
df['Cholesterol'].value_counts()
```

```python
df['RestingBP'].value_counts()
```

```python
df['RestingBP'].replace(0,df['RestingBP'].mean(),inplace=True)
```

```python
## convert to int
df['RestingBP']=df['RestingBP'].astype(int)
df['Cholesterol']=df['Cholesterol'].astype(int)
```
